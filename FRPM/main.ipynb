```python
import yfinance as yf

# Download AAPL closing prices from 2020-2026
prices = yf.download("AAPL", start="2020-01-01", end="2026-01-01", progress=False) ['Close'].squeeze()
```

# Financial Risk & Performance Metrics Analysis

### Comprehensive Portfolio Risk Assessment Using Python

```python
# Display the price data
print(prices)
```

```python
# Calculate daily returns (percentage change)
returns = prices.pct_change().dropna()
```

```python
returns
```

```python
# Calculate weekly returns using resampling
weekly_returns = prices.resample('W').last().pct_change().dropna()
```

```python
weekly_returns
```

```python
# Calculate monthly returns using resampling
monthly_returns = prices.resample('ME').last().pct_change().dropna()
```

```python
monthly_returns
```

```python
import empyrical as emp

# Sortino Ratio: Measures risk-adjusted return considering only downside deviation
emp.sortino_ratio(returns)
```

```python
# Sharpe Ratio: Measures excess return per unit of total risk (volatility)
emp.sharpe_ratio(returns)
```

```python
# Cumulative Returns: Total return over the entire period
emp.cum_returns_final(returns)
```

```python
# Annual Return: Compound annual growth rate (CAGR)
emp.annual_return(returns)
```

```python
# Annual Volatility: Standard deviation of annual returns
emp.annual_volatility(returns)
```

```python
# Aggregate Returns: Group returns by a given time period (monthly)
emp.aggregate_returns(returns, 'monthly')
```

```python
# Maximum Drawdown: Largest peak-to-trough decline
emp.max_drawdown(returns)
```

```python
# Downside Risk: Standard deviation of negative returns only
emp.downside_risk(returns)
```

```python
# Calmar Ratio: Annual return divided by maximum drawdown
emp.calmar_ratio(returns)
```

```python
# Define strategy (AAPL) and benchmark (SPY - S&P 500 ETF)
strategy = yf.download("AAPL", start="2020-01-01", end="2026-01-01", progress=False) ['Close'].squeeze()
benchmark = yf.download("SPY", start="2020-01-01", end="2024-01-01", progress=False) ['Close'].squeeze()
```

```python
# Calculate returns for both strategy and benchmark
strategy_returns = strategy.pct_change().dropna()
benchmark_returns = benchmark.pct_change().dropna()
```

```python
strategy_returns = strategy.pct_change().dropna()
benchmark_returns = benchmark.pct_change().dropna()
```

```python
# Align the two return series to ensure they have matching dates
strategy_returns, benchmark_returns = strategy_returns.align(benchmark_returns, join='inner')
```

```python
# Alpha and Beta: 
# Alpha - Excess return over benchmark (positive = outperformance)
# Beta - Sensitivity to market movements (1 = moves with market)
emp.alpha_beta(strategy_returns, benchmark_returns)
```

```python
import pandas as pd

# Custom function to calculate all key metrics for any given stock
def my_metric_calcutation(ticker, start_date='2020-01-01', end_date='2026-01-01'):
    """Calculate comprehensive risk and return metrics for a given stock ticker."""
    prices = yf.download(ticker, start=start_date, end=end_date, progress=False)['Close'].squeeze()
    returns = prices.pct_change().dropna()

    metrics = pd.DataFrame({
        'metric': ['sortino_ratio', 'sharpe_ratio', 'cum_returns_final', 'annual_return', 'annual_volatility', 'max_drawdown', 'downside_risk', 'calmar_ratio'],
        'value': [
            emp.sortino_ratio(returns),
            emp.sharpe_ratio(returns),
            emp.cum_returns_final(returns),
            emp.annual_return(returns),
            emp.annual_volatility(returns),
            emp.max_drawdown(returns),
            emp.downside_risk(returns),
            emp.calmar_ratio(returns)
        ]
    })
    return metrics
```

```python
# Example: Calculate metrics for Tesla (TSLA)
my_metric_calcutation('TSLA')
```
