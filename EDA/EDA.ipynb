```python
import os
import pandas as pd
import numpy as np

# Create the directory
os.makedirs('ecommerce_data', exist_ok=True)

# Create dummy orders data
orders_df = pd.DataFrame({
    'order_id': ['ord_1', 'ord_2'],
    'customer_id': ['cust_1', 'cust_2'],
    'order_status': ['delivered', 'delivered'],
    'order_purchase_timestamp': ['2023-01-01 10:00:00', '2023-01-02 11:00:00'],
    'order_approved_at': ['2023-01-01 10:30:00', '2023-01-02 11:30:00'],
    'order_delivered_carrier_date': ['2023-01-01 15:00:00', '2023-01-02 16:00:00'],
    'order_delivered_customer_date': ['2023-01-03 10:00:00', '2023-01-04 11:00:00'],
    'order_estimated_delivery_date': ['2023-01-05 10:00:00', '2023-01-06 11:00:00']
})
orders_df.to_csv('ecommerce_data/orders_dataset.csv', index=False)

# Create dummy items data
items_df = pd.DataFrame({
    'order_id': ['ord_1', 'ord_2'],
    'order_item_id': [1, 1],
    'product_id': ['prod_1', 'prod_2'],
    'seller_id': ['sell_1', 'sell_2'],
    'shipping_limit_date': ['2023-01-05', '2023-01-06'],
    'price': [100.0, 150.0],
    'freight_value': [10.0, 15.0]
})
items_df.to_csv('ecommerce_data/order_items_dataset.csv', index=False)

# Create dummy products data
products_df = pd.DataFrame({
    'product_id': ['prod_1', 'prod_2'],
    'product_category_name': ['electronics', 'books_media'],
    'product_name_length': [10, 12],
    'product_description_length': [100, 120],
    'product_photos_qty': [1, 2],
    'product_weight_g': [500, 300],
    'product_length_cm': [20, 15],
    'product_height_cm': [5, 2],
    'product_width_cm': [15, 10]
})
products_df.to_csv('ecommerce_data/products_dataset.csv', index=False)

# Create dummy customers data
customers_df = pd.DataFrame({
    'customer_id': ['cust_1', 'cust_2'],
    'customer_unique_id': ['uniq_1', 'uniq_2'],
    'customer_zip_code_prefix': ['12345', '67890'],
    'customer_city': ['New York', 'Los Angeles'],
    'customer_state': ['NY', 'CA']
})
customers_df.to_csv('ecommerce_data/customers_dataset.csv', index=False)

# Create dummy reviews data
reviews_df = pd.DataFrame({
    'review_id': ['rev_1', 'rev_2'],
    'order_id': ['ord_1', 'ord_2'],
    'review_score': [5, 4],
    'review_comment_title': ['Great', 'Good'],
    'review_comment_message': ['Fast', 'Nice'],
    'review_creation_date': ['2023-01-04', '2023-01-05'],
    'review_answer_timestamp': ['2023-01-05', '2023-01-06']
})
reviews_df.to_csv('ecommerce_data/order_reviews_dataset.csv', index=False)

print("Mock ecommerce_data created successfully.")
```

```python
# How much revenue did we generate in 2023 compared to 2022?
```

```python
order_items.head(2)
```

```python
orders.head(2)
```

```python
sales_data = pd.merge(left = order_items[['order_id','order_item_id', 'product_id', 'price']],
         right = orders[['order_id', 'order_status', 'order_purchase_timestamp', 'order_delivered_customer_date']],
         on = 'order_id')
```

```python
sales_data.head(2)
```

```python
sales_data['order_status'].unique()
```

```python
sales_delivered = sales_data[sales_data['order_status'] == 'delivered']
```

```python
sales_delivered['order_purchase_timestamp'] = pd.to_datetime(sales_delivered['order_purchase_timestamp'])
```

```python
sales_delivered['month'] = sales_delivered['order_purchase_timestamp'].apply(lambda timestamp:timestamp.month)
```

```python
sales_delivered['year'] = sales_delivered['order_purchase_timestamp'].apply(lambda timestamp:timestamp.year)
```

```python
sales_delivered
```

```python
sales_delivered_2023 = sales_delivered[sales_delivered['year']==2023]
sales_delivered_2022 = sales_delivered[sales_delivered['year']==2022]
```

```python
#total revenue in 2023
print(f"total_revenue in 2023: {sales_delivered_2023['price'].sum()}")
```

```python
# revenue compared to 2022
revenue_growth=(sales_delivered_2023['price'].sum()-sales_delivered_2022['price'].sum())/sales_delivered_2022['price'].sum()
print(f"{revenue_growth*100}%")
```

```python
# What is the month-over-month growth trend in 2023?
```

```python
monthly_growth = sales_delivered_2023[['price', 'month']].groupby('month')['price'].sum().pct_change()
```

```python
monthly_growth
```

```python
pct = monthly_growth.mean()
```

```python
print(f"Monthly growth in 2023: {100*pct}%")
```

```python
# average order value in 2023
```

```python
print(f"Average order value in 2023: {sales_delivered_2023.groupby('order_id')['price'].sum().mean()}")
```

```python
avg_order_2023 = sales_delivered_2023.groupby('order_id')['price'].sum().mean()
avg_order_2022 = sales_delivered_2022.groupby('order_id')['price'].sum().mean()

print(f"Compared to last year: {100*(avg_order_2023 - avg_order_2022)/(avg_order_2022)}%")
```

```python
# Total orders in 2023
print(f"Total number of orders in 2023: {sales_delivered_2023['order_id'].nunique()}")
```

```python
total_orders_2023 = sales_delivered_2023['order_id'].nunique()
total_orders_2022 = sales_delivered_2022['order_id'].nunique()

if total_orders_2022 > 0:
    growth = 100 * (total_orders_2023 - total_orders_2022) / total_orders_2022
    print(f"Compared to last year: {growth}%")
else:
    print("Compared to last year: N/A (No data for 2022)")
```
